```python
from IPython.display import clear_output
 
def mostra_taulell(taulell, diccionari_jugadors):
    n = len(taulell)
    print("    " + "   ".join(str(i) for i in range(n)))
    for fila in range(n):
        contingut = " | ".join(diccionari_jugadors[taulell[fila][columna]] for columna in range(n))
        print(f" {fila}  {contingut}")
        if fila < n-1:
            print("   " + "---+" *(n-1) + "---")
    print()
    return


def demana_posicio(taulell):
    while True:
        try:
            fila = int(input("Dona'm la fila on vols posar la peça (valor entre 0 i n)"))
            while fila < 0 or fila > (n-1):
                fila = int(input("Siusplau, dona'm be la fila on vols posar la peça (valor entre 0 i n)"))
        except:
            return None, None
        try:
            columna = int(input("Dona'm la columna on vols posar la peça (valor entre 0 i n)"))
            while columna < 0 or columna > (n-1):
                columna = int(input("Siusplau, dona'm be la fila on vols posar la peça (valor entre 0 i n)"))
        except:
            return None, None
        if taulell[fila][columna] != 0:
            print("Aquesta posició ja està ocupada. Torna-ho a intentar.")
            continue
        else:
            break

    return fila, columna

def actualitza_taulell(taulell, fila, columna, jugador):
    taulell[fila][columna] = jugador
    return taulell

def canvi_jugador(jugador):
    return jugador*(-1)

def comprovar_victoria(taulell,jugador):
    # Victòria per files
    for fila in range(0,n):
        suma = 0
        for columna in range(0,n):
            suma += taulell[fila][columna]
        if suma == n*jugador:
            return True
    # Victòria per columnes
    for columna in range(0,n):
        suma = 0
        for fila in range(0,n):
            suma += taulell[fila][columna]

        if suma == n*jugador:
            return True 
    # Victòria per diagonals
    suma = 0
    for diagonal in range(0,n):
        suma += taulell[diagonal][diagonal]            
    if suma == n*jugador:
        return True 
    # Victòria per diagonal
    suma = 0
    for diagonal in range(0,n):
        suma += taulell[diagonal][(n-1)-diagonal]            
    if suma == n*jugador:
        return True 
    
    return False


```

```python

jugador = 1
diccionari_jugadors = {
    0: " ",
    1: "X",
    -1: "O"}
n = int(input("Introdueix un nombre per la mida del taulell: "))

taulell=[[0 for num in range(n)] for i in range(n)]
condicio_victoria = False
comptador = 0

while not condicio_victoria and comptador < n*n:
    clear_output(wait=True)
    mostra_taulell(taulell,diccionari_jugadors)
    fila, columna = demana_posicio(taulell)
    if fila == None and columna == None:
        print("Has de donar un nombre natural entre 0 i n. Sortim del joc")
        break
    taulell = actualitza_taulell(taulell, fila, columna, jugador)
    condicio_victoria = comprovar_victoria(taulell, jugador)
    if condicio_victoria:
        clear_output(wait=True)
        mostra_taulell(taulell,diccionari_jugadors)
        print(f"Ha guanyat el jugador {diccionari_jugadors[jugador]}")
        break

    comptador += 1
    jugador = canvi_jugador(jugador)



```
